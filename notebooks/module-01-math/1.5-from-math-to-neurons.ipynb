````python
import numpy as np

# Step 0: Our input vector (e.g., three properties of a word)
x = np.array([1.5, -2.0, 0.5])

# Step 1: Weight Matrix (learned by the AI over time)
W = np.array([
    [0.1, 0.5, -0.1],
    [0.2, -0.3, 0.8],
    [0.9, 0.1, -0.4]
])

# Step 2: Bias Vector (also learned by the AI)
b = np.array([0.1, 0.2, 0.1])

# Step 3: Perform the calculation
# Multiply weights by input and add bias
linear_step = (W @ x) + b

# Step 4: Apply the Activation Function (ReLU)
# np.maximum(0, array) compares every element to 0
y = np.maximum(0, linear_step)

print(f"Input Vector (x): {x}")
print(f"Linear Result (Wx + b): {linear_step}")
print(f"Final Neuron Output (y): {y}")
````

**Exercise 2:** Why is the ReLU activation function important? What happens if we remove it and only use linear transformations?

<details>
<summary>Show solution</summary>

ReLU adds non-linearity, allowing neural networks to learn complex patterns. Without it:

```python
import numpy as np

x = np.array([1.0, 2.0])
W1 = np.array([[1.0, 2.0], [3.0, 4.0]])
b1 = np.array([0.5, 1.0])
W2 = np.array([[2.0, 3.0], [4.0, 5.0]])
b2 = np.array([0.2, 0.3])

# Without activation (just linear)
z1 = W1 @ x + b1
z2 = W2 @ z1 + b2

print(f"Two linear layers result: {z2}")
print("This is still just a linear transformation!")
print(f"It's equivalent to: W2 @ W1 @ x + (W2 @ b1 + b2)")
```

Without ReLU, stacking multiple layers is mathematically equivalent to having just one layer, so the network cannot learn complex patterns.
