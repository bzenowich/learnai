````python
import numpy as np

# 1. Let's create a tiny Embedding Matrix
# Each row represents a word from our vocabulary:
# [Apple, Orange, Toaster]
# Each vector has 3 dimensions (in real life, it's hundreds!)
embedding_matrix = np.array([
    [0.9, 0.1, 0.8], # Apple vector
    [0.8, 0.2, 0.7], # Orange vector
    [0.1, 0.9, 0.2]  # Toaster vector
])

# 2. Let's say our tokenizer gave us the ID for "Orange" (ID 1)
orange_id = 1

# 3. Look up the embedding vector
orange_vector = embedding_matrix[orange_id]

print(f"The embedding vector for 'Orange' (ID {orange_id}) is:")
print(orange_vector)

# Now, we have a mathematical vector that we can use 
# for dot products, matrix multiplication, and more!
````

**Exercise 1:** Create an embedding matrix with three 3D vectors for `cat`, `dog`, and `car`. Calculate the dot product between the `cat` and `dog` vectors. Which pair should have a higher dot product: cat-dog or cat-car?

<details>
<summary>Show solution</summary>

Create the embedding matrix and compute the dot product:

```python
import numpy as np
embedding_matrix = np.array([
    [0.7, 0.3, 0.1],  # cat
    [0.6, 0.5, 0.2],  # dog
    [0.2, 0.2, 0.9]   # car
])
cat_id = 0
dog_id = 1
cat_vector = embedding_matrix[cat_id]
dog_vector = embedding_matrix[dog_id]
dot_product = np.dot(cat_vector, dog_vector)
print(f"Cat embedding: {cat_vector}")
print(f"Dog embedding: {dog_vector}")
print(f"Dot product: {dot_product:.4f}")
```

Running this prints:
```text
Cat embedding: [0.7 0.3 0.1]
Dog embedding: [0.6 0.5 0.2]
Dot product: 0.5900
```

The cat-dog dot product should be higher than cat-car because cats and dogs are semantically similar animals, so their embeddings should be closer together in vector space.

**Exercise 2:** If an embedding matrix has shape `(3, 2)`, what does this tell you about the vocabulary size and embedding dimension?

<details>
<summary>Show solution</summary>

An embedding matrix with shape `(3, 2)` contains:
- **Vocabulary size:** 3 (number of rows = number of words/tokens)
- **Embedding dimension:** 2 (number of columns = dimensions per vector)

```python
import numpy as np
embedding_matrix = np.array([
    [0.9, 0.1],
    [0.8, 0.2],
    [0.1, 0.9]
])
print(f"Shape: {embedding_matrix.shape}")
print(f"Vocab size: {embedding_matrix.shape[0]}")
print(f"Embedding dimension: {embedding_matrix.shape[1]}")
```

Running this prints:
```text
Shape: (3, 2)
Vocab size: 3
Embedding dimension: 2
```

**Exercise 3:** In the lesson, we saw that similar words like "apple" and "orange" have very close embeddings. Why is this property important for an AI model's ability to understand language?

<details>
<summary>Show solution</summary>

When similar words have close embeddings, the AI model can generalize from one word to another. This means if the model has learned something about "apple", it can apply similar reasoning to "orange" without having seen every specific example. This property allows the model to understand new or rare words by recognizing they are semantically similar to words it has already learned about. It also enables the model to capture semantic relationships like "fruit-ness" or "smallness" across many related words in a single multidimensional space.
</details>
