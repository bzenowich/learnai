````python
import json

# Simulate an MCP Server that provides tools
class MCPServer:
    def __init__(self):
        self.tools = {
            "file_system": {"name": "file_system", "description": "Read/write files on disk"},
            "github": {"name": "github", "description": "Search and interact with GitHub repos"},
            "database": {"name": "database", "description": "Query SQL databases"}
        }
    
    def list_tools(self):
        return self.tools

# Simulate an MCP Client (the AI application)
class MCPClient:
    def __init__(self, server):
        self.server = server
    
    def get_available_tools(self):
        return self.server.list_tools()

# Simulate the LLM using tools
server = MCPServer()
client = MCPClient(server)

print("Available MCP Tools:")
for tool_name, tool_info in client.get_available_tools().items():
    print(f"  - {tool_info['name']}: {tool_info['description']}")
````

**1. What is the key advantage of MCP over each AI platform building its own tool protocol?**

MCP is a universal standard. Without it, a tool developer would need to write separate integrations for ChatGPT, Claude, Gemini, and every other LLM. With MCP, they write once and it works everywhere. This dramatically reduces development effort, increases compatibility, and creates an ecosystem of reusable, interoperable tools.

**2. How does MCP differ from RAG in terms of what information it provides?**

**RAG** retrieves *static text content* from documents (books, articles) and adds it to context. **MCP** is a *protocol* for two-way communication: the AI can request data or trigger actions, and the server responds dynamically. RAG is read-only retrieval; MCP enables bidirectional interaction with tools and services.

**3. Extend the MCPServer to support adding a custom tool.**

```python
class MCPServer:
    def __init__(self):
        self.tools = {
            "file_system": {"name": "file_system", "description": "Read/write files on disk"},
            "github": {"name": "github", "description": "Search and interact with GitHub repos"},
            "database": {"name": "database", "description": "Query SQL databases"}
        }
    
    def add_tool(self, tool_name, description):
        self.tools[tool_name] = {"name": tool_name, "description": description}
    
    def list_tools(self):
        return self.tools

server = MCPServer()
server.add_tool("slack", "Post and read Slack messages")

print("Available MCP Tools:")
```
