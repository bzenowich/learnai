````python
# A Level 1 AI has no history
def level_1_chat(query):
    # It just looks at the single query and responds
    if "moon" in query.lower():
        return "Neil Armstrong was the first person on the moon."
    else:
        return "I'm not sure what you're asking."

# First interaction
print(level_1_chat("Who walked on the moon?"))

# Second interaction (The AI has "forgotten" everything!)
print(level_1_chat("What year did that happen?")) 
````

**Exercise 2:** Why does Level 1 fail at answering follow-up questions? What information is lost between turns?

**Answer:** Level 1 loses all context between turns because it processes only the current query. When you ask "What year did that happen?" without mentioning the moon, the model has no way to know what "that" refers to. The entire conversation history exists only in the user's mind, not in the model's input. This stateless architecture is why we need memory (Level 2) for coherent conversations.
