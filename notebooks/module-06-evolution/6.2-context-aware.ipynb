````python
# A Level 2 AI has a "history" list to simulate memory
history = []

def level_2_chat(query):
    # 1. Add current query to history
    history.append(f"User: {query}")
    
    # 2. Build the giant context-aware prompt
    full_context = "\n".join(history)
    
    # 3. Simulate processing (The AI "reads" the whole context!)
    response = ""
    if "moon" in full_context.lower() and "year" in query.lower():
        response = "Assistant: That happened in 1969."
    elif "moon" in query.lower():
        response = "Assistant: Neil Armstrong was the first person on the moon."
    
    # 4. Add the AI's response to history too!
    history.append(response)
    return response

# First interaction
print(level_2_chat("Who walked on the moon?"))

# Second interaction (The AI "remembers" via context!)
print(level_2_chat("What year did that happen?"))
````

**Exercise 2:** What is the main limitation of simply concatenating conversation history? What happens as the conversation grows?

**Answer:** As conversation history grows, the concatenated prompt becomes increasingly long. This consumes more of the model's [context window](../glossary.md#context-window) (Module 4), leaving less room for new information or complex instructions. Eventually, old messages may be "forgotten" if they exceed the window limit. This is why [RAG systems](../module-05-rag-tools/5.2-rag.md) (Module 5) are needed—to retrieve only the most relevant past messages rather than including everything.
