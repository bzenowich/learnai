````python
def level_3_reasoning(query):
    # Phase 1: The Hidden Thought Process
    print("[Internal Thought]: The user wants to calculate a 2D dot product.")
    print("[Internal Thought]: Formula is (a1 * b1) + (a2 * b2).")
    
    # Phase 2: A deliberate calculation step
    a = [1, 2]
    b = [3, 4]
    result = (a[0] * b[0]) + (a[1] * b[1])
    
    print("[Internal Thought]: Result is 11. Wait, let me check... 1*3=3, 2*4=8, 3+8=11. Looks correct.")

    # Phase 3: The final answer to the user
    return f"The dot product of [1, 2] and [3, 4] is {result}."

print(level_3_reasoning("Calculate the dot product."))
````

**Exercise 2:** Why does [Chain-of-Thought](../glossary.md#chain-of-thought) reasoning work better for complex problems than direct answers?

**Answer:** Chain-of-Thought forces the model to break problems into smaller steps. By showing reasoning step-by-step, the model can catch its own mistakes (self-correction) and doesn't have to predict the final answer in one shot. This reduces error propagation—an error in step 1 can be caught before step 2. For complex math or logic problems, this explicit intermediate reasoning is often crucial for correctness.
