````python
# A simple autonomous agent loop
def level_4_agent(user_task):
    print(f"[Plan]: {user_task}")
    
    # 1. Step one: Research
    print("[Action]: Researching current data via Tool...")
    current_data = {"price": 50000} # Simulated tool result
    
    # 2. Step two: Thinking / Processing
    print(f"[Reflection]: Okay, price is {current_data['price']}. Now I can do the math.")
    
    # 3. Step three: Action (like writing a file)
    print("[Action]: Writing Python script to 'bitcoin_calc.py'...")
    
    # 4. Final delivery
    return "Task Complete: The script has been written and verified."

print(level_4_agent("Calculate my Bitcoin buy."))
````

```python
def tool_file_reader(filename):
    return f"Contents of {filename}: [simulated data]"

def tool_calculator(operation):
    if operation == "sum(10, 20)":
        return 30
    elif operation == "multiply(5, 4)":
        return 20
    return 0

def level_4_agent_with_tools(task):
    print(f"[Plan]: {task}")
    
    if "read" in task.lower():
        print("[Action]: Calling file_reader tool...")
        result = tool_file_reader("data.txt")
        print(f"[Tool Result]: {result}")
    
    if "calculate" in task.lower():
        print("[Action]: Calling calculator tool...")
        result = tool_calculator("sum(10, 20)")
        print(f"[Tool Result]: The sum is {result}")
    
    return "Task Complete"

print(level_4_agent_with_tools("Read file data.txt and calculate sum(10, 20)"))
```

**Exercise 2:** How does an [agent](../glossary.md#agent) differ from a regular Level 2 chatbot? What key capability does Level 4 add?

**Answer:** Level 2 chatbots can only output text; they cannot take actions in the world. An [agent](../glossary.md#agent) (Level 4) can call external tools and respond based on their results, creating a feedback loop. This allows agents to gather real-time information, modify files, execute code, and self-correct based on tool feedback—creating autonomous behavior beyond just text generation.
